```python
#add imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read csv file
data = pd.read_csv('Data_science_salaries.csv')
```

```python
#filter data to show experience levels and company size
updated_data = data[data['company_size'].isin(['S','M','L'])]
```

```python
#Use groupby - grouping by experience level and company size
salary_vs_size = updated_data.groupby(['company_size', 'experience_level'])['salary_in_usd'].mean()
```

```python
#shape data and columns
reshape = salary_vs_size.unstack()
reshape
```

```python
#create bar chart
ax = reshape.plot(kind = 'bar', figsize = (8,8))
plt.title('Average Salaries by Experience Level and Company Size')
plt.xlabel('Company Size')
plt.ylabel('Average Salary($USD)')
```
